```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

dataset = pd.read_csv('top500mangaMAL.csv')
```

```python
dataset.head()
```

```python
dataStatus = pd.DataFrame(dataset.loc[:, ['Status', 'Score']])
dataStatus = dataStatus.sort_values(by='Score', ascending='True')

dataStatus
```

```python
#plt.figure(figsize=(15,5))
sns.catplot(x='Score',y='Status', data=dataStatus, height = 10)
plt.title("Etat d'avancement des mangas du top 500 de MyAnimeList")
plt.xlabel('Score')
plt.ylabel('Status')
plt.show()
```
