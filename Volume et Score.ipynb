```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

dataset = pd.read_csv('top500mangaMAL.csv')
```

```python
dataset.head()
```

```python
dataVolumes = pd.DataFrame(dataset.loc[:, ['Volumes', 'Score']])
dataVolumes = dataVolumes.replace('Unknown', '')
dataVolumes = dataVolumes.sort_values(by='Volumes')
dataVolumes
```

```python
plt.figure(figsize=(15,10))
sns.scatterplot(x=dataVolumes['Volumes'],y=dataVolumes['Score'])
plt.title('Nuage de points - Nombres de volumes par rapport au score')
plt.xlabel('Volumes')
plt.ylabel('Score')
plt.show()
```
